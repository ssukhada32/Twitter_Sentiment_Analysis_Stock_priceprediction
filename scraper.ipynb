```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
query =['(#TCS) until:2016-04-30 since:2016-01-01',
'(#tcs) until:2016-04-30 since:2016-01-01',
'(#Tcs) until:2016-04-30 since:2016-01-01',
'(#TATA) until:2016-04-30 since:2016-01-01',
'(#tata) until:2016-04-30 since:2016-01-01',
'(#Tata) until:2016-04-30 since:2016-01-01',
'(#tata consultancy sevices) until:2016-04-30 since:2016-01-01',
'(#Tata Consultancy Services) until:2016-04-30 since:2016-01-01',
'(#TATA CONSULTANCY SERVICES) until:2016-04-30 since:2016-01-01',
'(#Tata_consultancy_services) until:2016-04-30 since:2016-01-01',
'(#Tata_Consultancy_Services) until:2016-04-30 since:2016-01-01',
'(#TATA_CONSULTANCY_SERVICES) until:2016-04-30 since:2016-01-01',
'(#tata_consultancy_services) until:2016-04-30 since:2016-01-01']
```

```python
q = [item.strip() for item in query]
print(q)
```

```python
tweets=[]
limit=0
```

```python
for i in q:
    limit = limit+ 5000
    for tweet in sntwitter.TwitterSearchScraper(i).get_items():    
   
        if len(tweets)==limit:
            break
        else:
            tweets.append([tweet.date,tweet.user.username,tweet.content,
                           tweet.user.created,tweet.url,tweet.user.location,tweet.user.friendsCount,
                           tweet.user.favouritesCount,tweet.media,tweet.replyCount,tweet.retweetCount,tweet.retweetedTweet,
                           tweet.user.followersCount,tweet.hashtags])
```

```python
df_TCS = pd.DataFrame(tweets,columns=['Date', 'User', 'Tweet','Tweet_created','Url','Location','Friends_count',
                                      'Favourites_Count','Media','Reply_count','Retweet_count','Retweeted_tweet',
                                      'Followers_count','Hashtags'])
print(df_TCS) 
```

```python
df_TCS.to_csv('TCS_2016_1.csv')
```
